```python

def get_text_elements(driver):

    text_element = driver.find_elements(By.CLASS_NAME, 'rounded-mxl')

    return text_element

import sqlite3

import sqlite3

def insert_into_db(db_file, table, data):
    """
    Insere dados em uma tabela específica de um banco de dados SQLite.

    :param db_file: Caminho para o arquivo do banco de dados SQLite
    :param table: Nome da tabela onde os dados serão inseridos
    :param data: Dicionário contendo os dados a serem inseridos (chave: coluna, valor: valor)
    """
    try:
        # Conectar ao banco de dados
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()
        
        # Criar a query SQL para inserção
        columns = ', '.join(data.keys())
        placeholders = ', '.join('?' for _ in data)
        sql = f'INSERT INTO {table} ({columns}) VALUES ({placeholders})'
        
        # Executar a query
        cursor.execute(sql, tuple(data.values()))
        
        # Commit as alterações
        conn.commit()
        
        print("Dados inseridos com sucesso!")
    
    except sqlite3.Error as e:
        print(f"Erro ao inserir dados: {e}")
    
    finally:
        if conn:
            conn.close()

def reset_table(db_file, table_name):
    """
    Remove todos os registros de uma tabela específica de um banco de dados SQLite.

    :param db_file: Caminho para o arquivo do banco de dados SQLite
    :param table_name: Nome da tabela a ser zerada
    """
    try:
        # Conectar ao banco de dados
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()

        # Deletar todos os registros da tabela
        cursor.execute(f"DELETE FROM {table_name}")
        
        # Resetar o contador de autoincremento (opcional)
        cursor.execute(f"DELETE FROM sqlite_sequence WHERE name='{table_name}'")
        
        # Commit as alterações
        conn.commit()
        
        print(f"Tabela '{table_name}' zerada com sucesso!")
    
    except sqlite3.Error as e:
        print(f"Erro ao zerar a tabela: {e}")
    
    finally:
        if conn:
            conn.close()
```

```python
import sqlite3
from sqlite3 import Error

def reset_table(db_file, table_name):
    """
    Remove todos os registros de uma tabela específica de um banco de dados SQLite.

    :param db_file: Caminho para o arquivo do banco de dados SQLite
    :param table_name: Nome da tabela a ser zerada
    """
    try:
        # Conectar ao banco de dados
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()

        # Deletar todos os registros da tabela
        cursor.execute(f"DELETE FROM {table_name}")
        
        # Commit as alterações
        conn.commit()

        # Verificar se a tabela sqlite_sequence existe e resetar o autoincremento se necessário
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name='sqlite_sequence';")
        if cursor.fetchone():
            cursor.execute(f"DELETE FROM sqlite_sequence WHERE name='{table_name}'")
            conn.commit()
        
        print(f"Tabela '{table_name}' zerada com sucesso!")
    
    except Error as e:
        print(f"Erro ao zerar a tabela: {e}")
    
    finally:
        if conn:
            conn.close()

# Exemplo de uso
db_path = 'games.db'
table_name = 'nintendo64'
reset_table(db_path, table_name)
```
